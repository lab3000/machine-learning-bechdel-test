### get bechdel test scores, using sand box discoveries

```python
bechdel_test_scores = {}
score = None
for i in range(19757):
    if as_a_thinketh[i].has_attr('id'):
        if as_a_thinketh[i]['id'][:5]=='movie':
            movie = as_a_thinketh[i].contents[0]
            
            score = None
            i_ahead = 1
            while score==None:
                if not not as_a_thinketh[i+i_ahead].contents:
                    if as_a_thinketh[i+i_ahead].contents[0].has_attr('alt'):
                        if '[[' in as_a_thinketh[i+i_ahead].contents[0]['alt']:
                            score = as_a_thinketh[i+i_ahead].contents[0]['alt']
                            score = score.strip('[')
                            score = score.strip(']')
                            score = int(score)
                i_ahead+=1
    if score != None:
        bechdel_test_scores[movie] = score
```
